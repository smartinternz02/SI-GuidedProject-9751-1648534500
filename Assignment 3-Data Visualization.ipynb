```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_0 = pd.read_csv("iris.csv")
data_1 = pd.read_csv("titanic.csv")
```

```python
data_0.info()
```

```python
data_1.info()
```

# Activity1: Line Graph

```python
plt.plot(np.arange(0, 150), data_0["sepal_length"].values, 'C0')
plt.plot(np.arange(0, 150), data_0["sepal_width"].values, 'C1')
```

# Activity2: scatter plot

```python
plt.scatter(np.arange(0,150), data_0["petal_length"].values, color="C2")
plt.scatter(np.arange(0,150), data_0["petal_width"].values, color="C3")
```

# Activity3: bar graph

```python
plt.bar(np.arange(0, 150), data_0["sepal_length"].values, color="b")
plt.bar(np.arange(0, 150), data_0["sepal_width"].values, color="r")
```

# Activity4: histogram

```python
plt.hist(data_0["sepal_width"].values, color="C6")
```

# Activity5: count plot

```python
sns.countplot(x="who", data=data_1)
```

# Activity6: heatmap

```python
flights = sns.load_dataset("flights")
flights = flights.pivot("month", "year", "passengers")
sns.heatmap(flights)
```

# Activity7: boxplot

```python
sns.boxplot(x=data_1["fare"].values)
```

# Activity8: pairplot

```python
sns.pairplot(data_0)
```

# Activity9: jointplot

```python
sns.jointplot(data=data_0, x="petal_length", y="petal_width")
```

# Activity10: Kernel density plot

```python
sns.kdeplot(data=data_0)
```
